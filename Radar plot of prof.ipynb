```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot') #ggplot, seaborn-bright, seaborn-deep, tableau-colorblind10, fivethirtyeight, fast

#plt.style.use('seaborn-bright')


#Data for visualization
subjects=['EPC','WSI','DP','DHC', 'DM', 'AI']
#subjects=['','','','', '', '']

base=[100,100,100,100,100,100]
all_professionals=[16,6,18,20,16, 16]

students=[39,26,64,53,45,54]
doctors=[53,13,94,63,48,46]
pathologists=[74,26,85,64,55,44]
radiologists=[47,41,91,59,48,43]
technicians=[55,7,63,42,42,32]




#Obtain Angles
angles=np.linspace(0,2*np.pi,len(subjects), endpoint=False)
print(angles)

#making full cycle
angles=np.concatenate((angles,[angles[0]]))
subjects.append(subjects[0])

base.append(base[0])
all_professionals.append(all_professionals[0])
students.append(students[0])
doctors.append(doctors[0])
pathologists.append(pathologists[0])
radiologists.append(radiologists[0])
technicians.append(technicians[0])


#basic chart

#fig=plt.figure(figsize=(6,6))
#ax=fig.add_subplot(polar=True)
#ax.plot(angles,alice)
#plt.show()


"""
#custom 1
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)
#basic plot
ax.plot(angles,all_professionals, 'o--', color='g', label='All Professionals')
#fill plot
ax.fill(angles, all_professionals, alpha=0.25, color='g')
#Add labels
ax.set_thetagrids(angles * 180/np.pi, all_professionals)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
"""



#custom 2
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(111, polar=True,)

ax.fill(alpha=0, color='white')

#Baseline to show the maximum area Plot
ax.plot(angles,base, 'o-', color='white', linewidth=1)
ax.fill(angles, base, alpha=0.5, color='white')

#all professionals Plot
#ax.plot(angles,all_professionals, 'o-', color='mediumspringgreen', linewidth=1, label='All Professionals')
#ax.fill(angles, all_professionals, alpha=0.25, color='mediumspringgreen')

#students Plot
ax.plot(angles,students, 'o-', color='deeppink', linewidth=1, label='Students')
ax.fill(angles, students, alpha=0.25, color='deeppink')

#doctors Plot
ax.plot(angles,doctors, 'o-', color='blueviolet', linewidth=1, label='Doctors')
ax.fill(angles, doctors, alpha=0.25, color='blueviolet')

#pathologists Plot
ax.plot(angles,pathologists, 'o-', color='cyan', linewidth=1, label='Pathologists')
ax.fill(angles, pathologists, alpha=0.25, color='cyan')

#radiologists Plot
ax.plot(angles,radiologists, 'o-', color='darkseagreen', linewidth=1, label='Radiologists') # darkseagreen
ax.fill(angles, radiologists, alpha=0.25, color='mediumspringgreen')

#technicians Plot
ax.plot(angles,technicians, 'o-', color='yellow', linewidth=1, label='Technicians')
ax.fill(angles, technicians, alpha=0.25, color='yellow')

#for all
plt.rcParams.update({'font.size': 22})
# Set the legend font size
#plt.rc('legend', fontsize=15)
# Set the axes labels font size
#plt.rc('axes', labelsize=16)
# Set the axes title font size
#plt.rc('axes', titlesize=16)
# Set the font size for x tick labels
#plt.rc('xtick', labelsize=16)

ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
