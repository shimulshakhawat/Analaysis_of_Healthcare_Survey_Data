```python
import plotly.express as px
import pandas as pd
  
    
    
excel_file="Sentiments Analaysis.xlsx"

df=pd.read_excel(excel_file, sheet_name='calculation', usecols = 'A:N')

print(df.shape)

df.head()

    
plot = px.scatter_ternary(df, a = 'Neutral', 
                          b = 'Positive',
                          c='Negative',
                         color='Sentiment',  size_max=15,) #color='Color trinary' or Sentiment
plot.show()


#df = px.data.iris()

#print(df.shape)

#df.head()
  
    
fig = px.scatter_ternary(df, a="Neutral", b="Positive", c="Negative",
    color="Sentiment", size="Size", size_max=15,
    color_discrete_map = {"Neutral": "turquoise", "Positive": "blue", "Negative":"deeppink"} )
fig.show()
    
"""    
plot = px.scatter_ternary(df, a = 'sepal_width', 
                          b = 'sepal_length',
                          c='petal_width',
                          color = 'species',
                          size = 'petal_length')
plot.show()
"""

```
