```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')


# Import Data
#original data
#excel_file="Potential of workforce.xlsx"
#df=pd.read_excel(excel_file, sheet_name='calculation', usecols = 'A:K')


#correlation increase updated data
excel_file="Potential of workforce.xlsx"
df=pd.read_excel(excel_file, sheet_name='calculation', usecols = 'A:Y')


#df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/mpg_ggplot2.csv")

print(df.shape)

df.head()

col_psy_pr=df.iloc[:,9]
col_dig_com=df.iloc[:,10]


# Create Fig and gridspec
fig = plt.figure(figsize=(12, 12), dpi= 80)
grid = plt.GridSpec(4, 4, hspace=0.5, wspace=0.2)

# Define the axes
ax_main = fig.add_subplot(grid[:-1, :-1])
ax_right = fig.add_subplot(grid[:-1, -1], xticklabels=[], yticklabels=[])
ax_bottom = fig.add_subplot(grid[-1, 0:-1], xticklabels=[], yticklabels=[])

# Scatterplot on main ax
#ax_main.scatter('displ', 'hwy', s=df.cty*4, c=df.manufacturer.astype('category').cat.codes, alpha=.9, data=df, cmap="tab10", edgecolors='gray', linewidths=.5)

#original
#ax_main.scatter('Psychological preparedness', 'Digital health competency', alpha=.9, data=df, color="mediumseagreen", edgecolors='black', linewidths=0.5)

ax_main.scatter('Psychological preparedness updated', 'Digital health competency updated', alpha=.9, data=df, color="mediumseagreen", edgecolors='black', linewidths=0.5)



# histogram on the right
#ax_bottom.hist(df.displ, 40, histtype='stepfilled', orientation='vertical', color='deeppink')
ax_bottom.hist(col_psy_pr, 40, histtype='stepfilled', orientation='vertical', color='deeppink')
ax_bottom.invert_yaxis()

#col_psy_pr=df.iloc[:,9]
#col_dig_com=df.iloc[:,10]

# histogram in the bottom
#ax_right.hist(df.hwy, 40, histtype='stepfilled', orientation='horizontal', color='deeppink')
ax_right.hist(col_dig_com, 40, histtype='stepfilled', orientation='horizontal', color='deepskyblue')

# Decorations
#ax_main.set(title='Scatterplot with Histograms \n displ vs hwy', xlabel='Psychological preparedness', ylabel='Digital health competency')
ax_main.set(title='Scatterplot with Histograms \n Psychological preparedness vs Digital health competency', xlabel='Psychological preparedness', ylabel='Digital health competency')

ax_main.title.set_fontsize(20)
for item in ([ax_main.xaxis.label, ax_main.yaxis.label] + ax_main.get_xticklabels() + ax_main.get_yticklabels()):
    item.set_fontsize(14)

xlabels = ax_main.get_xticks().tolist()
ax_main.set_xticklabels(xlabels)
plt.show()


#print(df.displ)
```
